````python
df = pd.read_csv("Hotel Reservations.csv")
df_1 = df.copy()
df_1.head()
````

````python
# 2ai Enter your solution here
class NumericTransformers:
    """This class is used to apply 'Sigmoid' and 'Cube root'
    transformations to the input data (Vector)."""
    
    def __init__(self, *, features: list[str]) -> None:
        self.features = features

    def sigmoid_transformation(self, x):
        """This applies sigmoid transformation to the input data.
        It returns a value between 0 and 1.
        sigmoid = 1 / (1 + exp^(-x))
        """
        x = x.copy()
        for feat in self.features:
          x[feat] = 1/(1+np.exp(-x[feat]))
        return x
      
    def cube_root_transformation(self, x):
        """This applies cube_root transformation to the input data.
        cube_root = np.cbrt(X)
        """
        x = x.copy()
        for feat in self.features:
          x[feat] = np.cbrt(x[feat])
        return x

features = ["no_of_adults", "arrival_month", "no_of_special_requests"]
temp = df_1[["no_of_adults", "arrival_month", "no_of_special_requests"]]

trans = NumericTransformers(features=features)

trans.sigmoid_transformation(x = temp).head()
````

````python
# 4a Enter your solution here
df = pd.read_csv("laptops.csv")
df_1 = df.copy()
df_1["SSD"].head()
````

````python
def conversion(int_value, conversion_rate=1024):
    if len(int_value) == 1:
        value = int(int_value) * conversion_rate
    else:
        value = int_value
    return value
````

````python
pattern = r"(\d{1,3}\s*[tg]b*\s*ssd)"
replace_value = "unk"
df["SSD(GB)"] = df["SSD"].str.extract(pattern, flags=re.I)
df["SSD(GB)"] = df["SSD(GB)"].str.split().str[0]
df["SSD(GB)"] = np.where(df["SSD(GB)"].isna(), replace_value, df["SSD(GB)"])
df["SSD(GB)"] = df["SSD(GB)"].apply(conversion)
df["SSD(GB)"].head()
````

````python
pattern = r"(\d{5,10})"

df["price"] = df["price"].str.replace(",", "").str.extract(pattern).astype(float)
df["price"].head()
````

````python
def clean_warranty(input_: str):
  """This is used to convert the values to years"""
  NUM_MONTHS = 12
  if input_.lower().__contains__("month"):
    value = round(float(input_.split()[0]) / NUM_MONTHS, 2)
  elif input_.lower().__contains__("year"):
    value = int(input_.lower().split()[0])
  else:
    value = input_
  return value
````

````python
from pandas.core.arrays.categorical import contains
pattern = r"(\d{1,2}\s*\w{4,6})"
df["warranty"] = df["warranty"].fillna(value = replace_value)
df["warranty(years)"] = df["warranty"].str.extract(pattern, flags=re.I).values.flatten()
df["warranty(years)"] = df["warranty(years)"].fillna(value = replace_value)
df["warranty(years)"] = np.where(df["warranty(years)"].str.lower().str.contains(r"(year|months)"),
                                 df["warranty(years)"], replace_value,)
df["warranty(years)"]
````

````python
df["warranty(years)"] = df["warranty(years)"].apply(clean_warranty)
df["warranty(years)"]
````

````python
pattern = r"(\(\d*\.?\s*\d*\s*\w{4}\))"  # (15.6 Inch)

# Extract the pattern. It returns a DataFrame which should be converted back
# to a Series by flattening it.
df["Display(inch)"] = df["Display"].str.extract(pattern, flags=re.I).values.flatten()
df["Display(inch)"] = (
    df["Display(inch)"].str.lower().str.replace(r"[()inch]", "", regex=True)
)
df["Display(inch)"] = df["Display(inch)"].fillna(value=replace_value)
df["Display(inch)"].unique()
````
